```python
# TASK 2

class Assassin:

  count = 0

  def __init__(self,name,rate):
    self.name = name
    self.rate = rate
    Assassin.count += 1
  
  @classmethod
  def failureRate(cls,name,failure_rate):
    var = cls(name, 100 - failure_rate)
    return var
  
  @classmethod
  def failurePercentage(cls,name,failure_percent):
    var = cls(name, 100 - failure_percent)
    return var

  def printDetails(self):
    print("Name:", self.name)
    print("Success rate: " + str(self.rate) + "%")
    print("Total number of Assassin:", Assassin.count)


# Write your code here
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
# TASK 3

class Passenger:

  count = 0

  def __init__(self,name):
    Passenger.count +=1
    self.name = name

  def set_bag_weight(self,weight,fare=450):
    self.weight = weight
    self.fare = fare
    if self.weight <= 20 :
      return self.fare
    elif self.weight >= 21 and self.weight <= 50 :
      self.fare = self.fare + 50
      return self.fare
    elif self.weight > 50 :
      self.fare = self.fare + 100
      return self.fare

  def printDetail(self):
    print("Name:", self.name)
    print("Bus Fare:", self.fare, "taka")


# Write your code here
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
# TASK 4

class Travel:

  count = 0

  def __init__(self,source,destination,time=1):
    Travel.count += 1
    self.source = source
    self.destination = destination
    self.time = time

  def set_time(self,time):
    self.time = time

  def set_source(self,source):
    self.source = source

  def set_destination(self,destination):
    self.destination = destination

  def display_travel_info(self):
    var = "Source: " + self.source + "\n" + "Destination:" + self.destination + "\n" + "Flight Time:" + str(self.time) + ":00"
    return var


# Write your code here
print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
# TASK 5

class Employee:
  def __init__(self,name,workingPeriod):
    self.name = name
    self.workingPeriod = workingPeriod

  @classmethod
  def employeeByJoiningYear(cls,name,year):
    var = cls(name, 2021 - year)
    return var

  @staticmethod
  def experienceCheck(working_period,gender):
    if working_period < 3 and gender == "male" :
      var = "He is not experienced"
      return var
    else :
      out = "She is experienced"
      return out


# Write your code here
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
# TASK 6

class Laptop:

  laptopCount = 0

  def __init__(self,name,no):
    self.name = name
    self.count = no
    Laptop.laptopCount += no
  
  @staticmethod
  def advantage():
    print("Laptops are portable")

  @classmethod
  def resetCount(cls):
    cls.laptopCount = 0


# Write your code here
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
# TASK 7

class Cat:

  Number_of_cats=0
  
  def __init__(self,name,action):
    self.name = name
    self.action = action
    Cat.Number_of_cats += 1

  def changeColor(self,color):
    self.name = color

  @classmethod
  def no_parameter(cls):
    return cls(name = "White", action = "sitting")

  @classmethod
  def first_parameter(cls,name2):
    return cls(name2, action = "sitting")

  @classmethod
  def second_parameter(cls,action2):
    return cls(name = "Grey", action = action2)

  def printCat(self):
    print(self.name, "cat is", self.action)


# Write your code here
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
# TASK 8

import math

class Cylinder:
  
  radius = 5
  height = 18

  def __init__(self,radius,height):
    self.radius=radius
    self.height=height
    print("Default radius=" + str(Cylinder.radius) + " and height=" + str(Cylinder.height) + ".")
        
    Cylinder.radius = self.radius
    Cylinder.height = self.height
    print("Updated radius=" + str(self.radius) + " and height=" + str(self.height) + ".")

  @classmethod
  def swap(cls,new_radius,new_height):
    return cls(new_height,new_radius)

  @classmethod
  def changeFormat(cls,st):
    radius,height=st.split("-")
    return cls(float(radius),float(height))

  @staticmethod
  def area(radius,height):
    print("Area:", (2*math.pi*Cylinder.radius*Cylinder.radius)+(2*math.pi*Cylinder.radius*Cylinder.height))

  @staticmethod
  def volume(radius,height):
    print("Volume:", math.pi*(Cylinder.radius)*Cylinder.radius*Cylinder.height)


# Write your code here
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
# TASK 10

class SultansDine:
  num_of_branch = 0
  total_sell = 0
  food_list = []

  def __init__(self,branch_name):
    self.branch_name = branch_name
    SultansDine.num_of_branch += 1
    self.branch_sell = 0

  def branchInformation(self):
    print("Branch Name:", self.branch_name)
    print("Branch sell:", self.branch_sell, "Taka")

  def sellQuantity(self,quantity):
    if quantity < 10 :
      self.branch_sell = quantity * 300
    elif quantity < 20 :
      self.branch_sell = quantity * 350
    else :
      self.branch_sell = quantity * 400

    SultansDine.total_sell += self.branch_sell
    SultansDine.food_list.append(self)

  @classmethod
  def details(cls):
    print("Total Number of branch(s):", cls.num_of_branch)
    print("Total sell:", cls.total_sell, "Taka")
    
    for item in cls.food_list:
      print("Branch Name:", item.branch_name + "," + " Branch Sell: " + str(item.branch_sell) + " Taka")
      print("Branch consists of total sell's: " + "{:.2f}".format(item.branch_sell / cls.total_sell * 100) + "%")


# Write your code here
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
