```python
# TASK 1

class RealNumber:
  def __init__(self, r=0):
    self.__realValue = r
  def getRealValue(self):
    return self.__realValue
  def setRealValue(self, r):
    self.__realValue = r
  def __str__(self):
    return 'RealPart: '+str(self.getRealValue())

# Code

class ComplexNumber(RealNumber):
  def __init__(self,real=1,img=1):
    super().__init__(float(real))
    self.img_value = float(img)

  def __str__(self):
    output = super().__str__() + "\n" + "ImaginaryPart: " + str(self.img_value)
    return output


cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
# TASK 2

class RealNumber:
  def __init__(self, number=0):
    self.number = number
  def __add__(self, anotherRealNumber):
    return self.number + anotherRealNumber.number
  def __sub__(self, anotherRealNumber):
    return self.number - anotherRealNumber.number
  def __str__(self):
    return str(self.number)

# Code

class ComplexNumber(RealNumber):
  def __init__(self,real,img):
    if type(real) == int :
    #if type(real) == isinstance(real,int) :
      super().__init__(real)
    else :
      super().__init__(real.number)

    self.img_value = img

  def __str__(self):
    if self.img_value > 0 :
      return super().__str__() + " + " + str(self.img_value) + "i"
    else :
      return super().__str__() + " - " + str((-1) * self.img_value) + "i"

  def __add__(self,other):
    real = super().__add__(other)
    img = self.img_value + other.img_value
    obj = ComplexNumber(real,img)
    return obj

  def __sub__(self,other):
    real = super().__sub__(other)
    img = self.img_value - other.img_value
    obj = ComplexNumber(real,img)
    return obj


r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
# TASK 3

class Account:
  def __init__(self, balance):
    self._balance = balance
  def getBalance(self):
    return self._balance

# Code

class CheckingAccount(Account):
  numberOfAccount = 0

  def __init__(self,balance=0.0):
    super().__init__(balance)
    CheckingAccount.numberOfAccount += 1

  def __str__(self):
    if self._balance == 0 :
      return "Account Balance: " + str(self._balance)
    else :
      ac_balance = "Account Balance: " + str("{:.2f}".format(self._balance))
      return ac_balance


print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
print(CheckingAccount())
print(CheckingAccount(100.00))
print(CheckingAccount(200.00))
print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
```

```python
# TASK 5

class Exam:
  def __init__(self,marks):
    self.marks = marks
    self.time = 60
  def examSyllabus(self):
    return "Maths , English"
  def examParts(self):
    return "Part 1 - Maths\nPart 2 - English\n"

# Code

class ScienceExam(Exam):
  def __init__(self,marks,time,*sub):
    super().__init__(marks)
    self.time = time
    self.sub = sub

  def examSyllabus(self):
    var = super().examSyllabus()
    for sub in self.sub :
      var += ", " + sub
    return var

  def examParts(self):
    out = super().examParts()
    for index in range(len(self.sub)):
      if index == len(self.sub)-1:
        out += "Part " + str(index+3) + " - " + self.sub[index]
      else :
        out += "Part " + str(index+3) + " - " + self.sub[index] + "\n"
    return out

  def __str__(self):
    obj = "Marks: " + str(self.marks) + " Time: " + str(self.time) + " minutes Number of Parts: " + str(len(self.sub)+2)
    return obj


engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
# TASK 6

class Shape3D:
  pi = 3.14159
  def __init__(self, name = 'Default', radius = 0):
    self._area = 0
    self._name = name
    self._height = 'No need'
    self._radius = radius
  def calc_surface_area(self):
    return 2 * Shape3D.pi * self._radius
  def __str__(self):
    return "Radius: "+str(self._radius)

# Code

class Sphere(Shape3D):
  def __init__(self,name,radius):
    super().__init__(name,radius)
    obj = "Shape name: " + self._name + ", Area Formula: 4 * pi * r * r"
    print(obj)

  def calc_surface_area(self):
    self._area = 4 * Shape3D.pi * self._radius * self._radius

  def __str__(self):
    output = super().__str__() + ", Height: " + str(self._height) + "\n" + "Area: " + str(self._area)
    return output

class Cylinder(Shape3D):
  def __init__(self,name,radius,height):
    super().__init__(name,radius)
    self._height = height
    print("Shape name: " + self._name + ",  Area Formula: 2 * pi * r * (r + h)")

  def calc_surface_area(self):
    self._area = 2 * Shape3D.pi * self._radius * (self._radius + self._height)

  def __str__(self):
    output = super().__str__() + ", Height: " + str(self._height) + "\n" + "Area: " + str(self._area)
    return output


sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
# TASK 7

class PokemonBasic:
  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type
  def get_type(self):
    return 'Main type: ' + self.type
  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'
  def __str__(self):
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness

# Code

class PokemonExtra(PokemonBasic):
  def __init__(self,name = 'Default', hp = 0, weakness = 'None', type = 'Unknown',s_type=None,other_move=()):
    super().__init__(name,hp,weakness,type)
    self.s_type = s_type
    self.other_move = other_move

  def get_type(self):
    if self.s_type != None :
      second = super().get_type() + ", Secondary type: " + self.s_type
      return second
    else :
      return super().get_type()

  def get_move(self):
    obj = ""
    for move in range(len(self.other_move)):
      if move != len(self.other_move)-1 :
        obj += self.other_move[move] + ", "
      else :
        obj += self.other_move [move]
    
    if len(self.other_move) != 0 :
      other = super().get_move() + "\n" + "Other move: " + obj
      return other
    else :
      return super().get_move()


print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water','Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water','Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
# TASK 9

class Pokemon:
  def __init__(self, p):
    self.pokemon = p
    self.pokemon_type = "Needs to be set"
    self.pokemon_weakness = "Needs to be set"
  def kind(self):
    return self.pokemon_type
  def weakness(self):
    return self.pokemon_weakness
  def what_am_i(self):
    print("I am a Pokemon.")

# Code

class Pikachu(Pokemon):
  def __init__(self,p="Pikachu"):
    super().__init__(p)
    self.pokemon_type = "Electric"
    self.pokemon_weakness = "Ground"

  def what_am_i(self):
    super().what_am_i()
    print("I am Pikachu.")

class Charmander(Pokemon):
  def __init__(self,p="Charmander"):
    super().__init__(p)
    self.pokemon_type = "Fire"
    self.pokemon_weakness = "Water, Ground and Rock"

  def what_am_i(self):
    super().what_am_i()
    print("I am Charmander.")


pk1 = Pikachu()
print("Pokemon:", pk1.pokemon)
print("Type:", pk1.kind())
print("Weakness:", pk1.weakness())
pk1.what_am_i()
print("========================")
c1 = Charmander()
print("Pokemon:", c1.pokemon)
print("Type:", c1.kind())
print("Weakness:", c1.weakness())
c1.what_am_i()
```

```python
# TASK 10

class Department:
  def __init__(self, s):
    self.semester = s
    self.name = "Default"
    self.id = -1
  def student_info(self):
    print("Name:", self.name)
    print("ID:", self.id)
  def courses(self, c1, c2, c3):
    print("No courses Approved yet!")

# Code

class CSE(Department):
  def __init__(self,name,id,s):
    super().__init__(s)
    self.name = name
    self.id = id

  def student_info(self):
    return super().student_info()

  def courses(self, c1, c2, c3):
    self.c1 = c1
    self.c2 = c2
    self.c3 = c3
    print("Courses Approved to this CSE student in " + self.semester + " semester : " + "\n" + self.c1 + "\n" + self.c2 + "\n" + self.c3)

class EEE(Department):
  def __init__(self,name,id,s):
    super().__init__(s)
    self.name = name
    self.id = id

  def student_info(self):
    return super().student_info()

  def courses(self, c1, c2, c3):
    self.c1 = c1
    self.c2 = c2
    self.c3 = c3
    print("Courses Approved to this EEE student in " + self.semester + " semester : " + "\n" + self.c1 + "\n" + self.c2 + "\n" + self.c3)


s1 = CSE("Rahim", 16101328,"Spring2016")
s1.student_info()
s1.courses("CSE110", "MAT110", "ENG101")
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.student_info()
s2.courses("Mat110", "PHY111", "ENG101")
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.student_info()
s3.courses("CSE111", "PHY101", "MAT120")
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.student_info()
s4.courses("EEE201", "PHY112", "MAT120")
```
