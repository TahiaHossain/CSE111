```python
# TASK 1

class PlayerEarning:
  def __init__(self,name):
    self.name = name

  def calculateTotal(self,earning_without_bonus,goal=0):
    self.earning_without_bonus = earning_without_bonus
    self.goal = goal
    if self.goal == 0 :
      self.bonus = 0
    elif self.goal > 30 :
      self.bonus = (5/100) * earning_without_bonus + 10000
    else : 
      self.bonus = (5/100) * earning_without_bonus


  def printDetails(self):
    print("Player Name:", self.name)
    print("Player Season Earning without bonus:", self.earning_without_bonus)
    print("Bonus:", self.bonus)
    print("Player Season Earning After Bonus:", self.earning_without_bonus + self.bonus)

# Write your code here

print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()

print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()

print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
# TASK 3

class Bird:
  def __init__(self,name,condition=False):
    self.name = name
    self.condition = condition
    self.bird_type = "Flightless Birds"

  def fly(self):
    if self.condition == False :
      print(self.name, "can not fly")
    else :
      print(self.name, "can fly")

  def setType(self,bird_type):
    self.bird_type = bird_type

  def printDetail(self):
    print("Name:", self.name)
    print("Type:", self.bird_type)


ostrich = Bird('Ostrich')
duck = Bird("Duck", True)
owl = Bird('Owl', True)
print("###########################")
ostrich.fly()
duck.fly()
owl.fly()
duck.setType('Water Birds')
owl.setType('Birds of Prey')
print("=========================")
ostrich.printDetail()
print("=========================")
duck.printDetail()
print("=========================")
owl.printDetail()
```

```python
# TASK 4

class Account:
  count = 0
  def __init__(self,name,age,job,amount):
    self.name = name
    self.age = age
    self.job = job
    self.amount = amount
    Account.count += 1

  def addMoney(self,add):
    self.add = add
    self.amount += self.add

  def withdrawMoney(self,withdraw):
    self.withdraw = withdraw
    if self.withdraw > self.amount :
      self.amount = self.amount
    else :
      self.amount -= self.withdraw

  def printDetails(self):
    print("Name:", self.name)
    print("Age:", self.age)
    print("Occupation:", self.job)
    print("Total Amount:", self.amount)

# Write your code here
 
print('No of account holders:', Account.count)
print("=========================")
p1 = Account("Abdul", 45, "Service Holder", 500000)
p1.addMoney(300000)
p1.printDetails()
print("=========================")
p2 = Account("Rahim", 55, "Businessman", 700000)
p2.withdrawMoney(700000)
p2.printDetails()
print("=========================")
p3 = Account("Ashraf", 62, "Govt. Officer", 200000)
p3.withdrawMoney(250000)
p3.printDetails()
print("=========================") 
print('No of account holders:', Account.count)
```
